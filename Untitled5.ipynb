```python
#50_Startups
import pandas as pd
url='/50_Startups.csv'
data=pd.read_csv(url)
data.head()

import matplotlib.pyplot as plt
data.hist()
plt.show()

correlation=data.corr()
print(correlation)
```

```python

import seaborn as sns
sns.heatmap(correlation,annot=True)

sns.pairplot(data)
     
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
#learning data
model.fit(X=data.loc[:,['R&D Spend','Administration','Marketing Spend']],y=data['Profit'])
print('intercept la: ',model.intercept_)
print('coeffient la: ',model.coef_)
print('do chinh xac: ',model.score(X=data.loc[:,['R&D Spend','Administration','Marketing Spend']],y=data['Profit']))

#bac 3
model=linear_model.LinearRegression()
X=pd.DataFrame([data['Marketing Spend'],data['Marketing Spend']**2,data['Marketing Spend']**3]).T #chuyen vi hang thanh cot
model.fit(X,y=data['Profit'])
print("Intercept bac 3 cua mo hinh:", model.intercept_)
print("Coeffient bac 3 cua mo hinh:", model.coef_)
print('Do chinh xac bac 3 cua mo hinh:',model.score(X,y=data['Profit'])) #do chinh xac pt 3
```

```python
import numpy as np
range= np.arange(0,500000,1)
X_range=pd.DataFrame([range,range**2,range**3]).T #pt bac 3
y_pred=model.predict(X=X_range)
data.plot(kind='scatter',x='Marketing Spend',
          y='Profit',figsize=(9,9),color='black',
          xlim=(0,500000))
plt.plot(range,y_pred,color='blue')
```
